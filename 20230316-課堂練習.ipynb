匯入所需的套件，需要擺在程式的開頭。

```python
import numpy as np
import matplotlib.pyplot as plt
```

先定義定義簡單線性回歸模型訓練的function，依據python語法，def擺在最前面，return擺在最後面。

使用list紀錄每次迭代的值，因此需再迭代前先宣告
在迭代前先設定權重跟偏差

使用迴圈進行迭代的計算
    將權重跟偏差代入模型，可以得到預測值以及誤差
    使用梯度下降法計算並更新權重跟偏差值
    將loss、權重、偏差的值分別加到對應的list裡面

```python
def linear_regression(X,y,lr=0.1,n_iterations=10): #定義linear_regression這個function

    loss_history=[]  #存放每一次iteration的loss值
    b0_history=[]  #存放每一次iteration的偏差(截距)值
    w1_history=[]  #存放每一次iteration的權重(斜率)值

    b0=0  #偏差(截距)初始值設為0
    w1=0  #權重(斜率)初始值設為0
    
    for i in range(n_iterations):
        y_pred=b0+w1*X  #定義簡單回歸模型
        loss = np.mean((y_pred-y)**2)  #定義loss function(這裡採用MSE)

        #======定義gradient descent 梯度下降法計算步驟=====#
        grad_b0=np.mean(2*(y_pred -y)) # b0斜率
        grad_w1=np.mean(2*X*(y_pred -y)) # w1斜率
        b0 = b0 - lr*grad_b0
        w1 = w1 - lr*grad_w1
        #======定義gradient descent 梯度下降法計算步驟=====#


        loss_history.append(loss)
        b0_history.append(b0)
        w1_history.append(w1)   

    return b0,w1,loss_history,b0_history,w1_history
```

隨機生成訓練數據，此部分可以跟上面的def function互換

```python
#隨機生成一組訓練數據
np.random.seed(0)
n_samples = 100
X = np.random.randn(n_samples)  #隨機生成的訓練數據
y = 2*X+1+np.random.randn(n_samples)  #隨機生成的label數據
```

呼叫簡單線性回歸模型訓練的function，將隨機生成的訓練數據丟入模型訓練。

```python
w0,w1,loss_history,w0_history,w1_history=linear_regression(X,y)
```

將結果繪製出來，因為需要等數據將模型訓練出來，因此需要擺在最後面

```python
plt.figure(figsize=(10,5))
plt.plot(loss_history)
plt.xlabel('Iteration')
plt.ylabel('MSE Loss')
plt.title('Loss Trend')

plt.figure(figsize=(10,5))
plt.plot(w0_history)
plt.xlabel('Iteration')
plt.ylabel('w0')
plt.title('w0 Trend')

plt.plot(w1_history)
plt.xlabel('Iteration')
plt.ylabel('w1')
plt.title('w1 Trend')

plt.figure(figsize=(10,5))
plt.scatter(X,y)
plt.plot(X,w0+w1*X,color='r')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
```
